```python
def pnf_signal(df, plot_chart=False, min_period=100):
    df.index = pd.to_datetime(df.index).strftime('%Y-%m-%d')
    df = df.dropna()
    if df.shape[0] > min_period:
        i._symbol = df.name
        dat = OrderedDict()
        for index, value in df.items():
            d = OrderedDict()
            d['Date'] = index
            d['Open'] = Decimal(value)
            d['High'] = Decimal(value)
            d['Low'] = Decimal(value)
            d['Close'] = Decimal(value)
            d['Volume'] = 0
            dat[index] = d
        c = PFChart(i,
                    box_size=0.0325,
                    duration=duration,
                    reversal=3,
                    method='c',
                    style=True,
                    trend_lines=True)
        c.security.historical_data = dat
        c.create_chart()
        if c.current_signal == 'buy' and c.current_status == 'bull confirmed':
            signal = 2
        elif c.current_signal == 'buy' and c.current_status == 'bull correction':
            signal = 1
        elif c.current_signal == 'sell' and c.current_status == 'bear correction':
            signal = -1
        elif c.current_signal == 'sell' and c.current_status == 'bear confirmed':
            signal = -2
        else:
            signal = 0
        if plot_chart:
            print(c.chart)
    else:
        signal = np.nan
    return (signal)
```

```python
def _color_red_or_green(val):
    if val == 2:
        color = 'Green'
    elif val == 1:
        color = 'YellowGreen'
    elif val == -1:
        color = 'Tomato'
    elif val == -2:
        color = 'OrangeRed'
    else:
        color = 'white'
    return 'background-color: %s' % color
```
